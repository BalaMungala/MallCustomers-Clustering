```python
# importing dataset

import pandas as pd
dataset=pd.read_csv("I://3//Github-upload//cluster//Mall_Customers.csv")
x=dataset.iloc[:,[3,4]].values
```

```python
# finding number of clusters

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
#within cluster sum of squares
wcss=[]
for i in range(1,11):
        kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
        kmeans.fit(x)
        wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)        
plt.title("Elbow Method")
plt.xlabel("Number of custers")
plt.xlim(1,10)
plt.ylabel("wcss")
plt.savefig("I://3//Github-upload//cluster//em.png")
plt.show()
        
```

```python
# clustering and predicting 

kmeans=KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(x)
```

```python
# visualization

plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],c='r',label='Careful')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],c='g',label='Standard')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],c='b',label='Target')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],c='c',label='Careless')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],c='m',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c="y",label="centroids")
plt.title("clusters of clients")
plt.legend()
plt.xlabel("Annual Income")
plt.ylabel("Spending score")
plt.savefig("I://3//Github-upload//cluster//coc.png")
plt.show()
```
