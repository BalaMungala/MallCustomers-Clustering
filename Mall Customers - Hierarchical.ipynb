```python
# Importing dataset

import pandas as pd
dataset=pd.read_csv("I://3//Github-upload//cluster//Mall_Customers.csv")
x=dataset.iloc[:,[3,4]].values
```

```python
# Dendrogram

import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title("Dendogram")
plt.xlabel("customers")
plt.ylabel("Distances")
plt.savefig("I://3//Github-upload//cluster//dog.png")
plt.show()
```

```python
# clustering and predicting

from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,linkage='ward',affinity='euclidean')
y_hc=hc.fit_predict(x)
```

```python
# visualization

plt.scatter(x[y_hc==0,0],x[y_hc==0,1],c='r',label='Careful')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],c='g',label='Standard')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],c='b',label='Target')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],c='c',label='Careless')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],c='m',label='Sensible')
plt.title("clusters of clients")
plt.legend()
plt.xlabel("Annual Income")
plt.ylabel("Spending score")
plt.savefig("I://3//Github-upload//cluster//hcc.png")
plt.show()
```
